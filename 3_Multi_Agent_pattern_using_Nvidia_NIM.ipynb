#Usecase:product Research and Report Crew Agent

```python
# Define your agents
researcher = Agent(
  role='Researcher',
  goal='Conduct foundational research on the {product}',
  backstory='An experienced researcher with a passion for uncovering insights',
  llm=llm,

)
analyst = Agent(
  role='Data Analyst',
  goal='Analyze research findings',
  backstory='A meticulous analyst with a knack for uncovering patterns',
  llm=llm
)
writer = Agent(
  role='Writer',
  goal='Draft the final report',
  backstory='A skilled writer with a talent for crafting compelling narratives',
  llm=llm
)

# Define your tasks
research_task = Task(
  description='Gather relevant data...',
  agent=researcher,
  expected_output='Raw Data'
)
analysis_task = Task(
  description='Analyze the data...',
  agent=analyst,
  expected_output='Data Insights'
)
writing_task = Task(
  description='Compose the report...',
  agent=writer,
  expected_output='Final Report'
)
```

```python
from crewai import Process
```

```python
# Form the crew with a sequential process
report_crew = Crew(
  agents=[researcher, analyst, writer],
  tasks=[research_task, analysis_task, writing_task],
  process=Process.sequential
)
inputs = {
        'product':'Green tech software product Eco-Chain'


    }
```

```python

from crewai_tools import SerperDevTool
from crewai import Agent, Task, Crew, LLM # Importing LLM after upgrading crewai
from typing import Callable


# Initialize Tools
search_tool = SerperDevTool()

# Define a function to wrap the search_tool's run method
def search_tool_func(query: str) -> str:
    """
    This function is a wrapper for the SerperDevTool's run method.
    It allows the tool to be used with the Agent class.
    """
    return search_tool.run(query)


# Define the search tool with name, func, and description
search_tool_agent = {
    "name": "Search",
    "func": search_tool_func,
    "description": "Use this tool to search the internet for information."
}

researcher = Agent(
  role='Researcher',
  goal='Conduct foundational research on the {product}',
  backstory='An experienced researcher with a passion for uncovering insights',
  llm=llm,
  tools=[search_tool_agent] # Pass the tool as a dictionary
)
analyst = Agent(
  role='Data Analyst',
  goal='Analyze research findings',
  backstory='A meticulous analyst with a knack for uncovering patterns',
  llm=llm,
  tools=[search_tool_agent] # Pass the tool as a dictionary

)
writer = Agent(
  role='Writer',
  goal='Draft the final report',
  backstory='A skilled writer with a talent for crafting compelling narratives',
  llm=llm,
  tools=[search_tool_agent] # Pass the tool as a dictionary
)

# Define your tasks
research_task = Task(
  description='Gather relevant data...',
  agent=researcher,
  expected_output='Raw Data'
)
analysis_task = Task(
  description='Analyze the data...',
  agent=analyst,
  expected_output='Data Insights'
)
writing_task = Task(
  description='Compose the report...',
  agent=writer,
  expected_output='Final Report'
)
```
